```python
from langchain_community.document_loaders import PyPDFLoader

# 로더 생성
loader = PyPDFLoader("./data/Demian.pdf")

# 1. 간단하게 전체 로드 (리스트 형태)
docs = loader.load()

# 데이터 구조 확인
print(docs[0].page_content)
print(docs[60].metadata)
```

```python
len(docs)
```

```python
import re


def clean_docs(docs):
    """특수문자, 개행문자 교체 및 양끝단 공백 제거"""    
    cleaned_docs = re.sub("\n", " ", docs)
    cleaned_docs = re.sub(r"[^a-zA-Z가-힣,.?!0-9\s]", " ", cleaned_docs)
    cleaned_docs = re.sub(r"\s+", " ", cleaned_docs).lower().strip()
    return cleaned_docs

for doc in docs:
    doc.page_content = clean_docs(doc.page_content)

print(docs[3].page_content)
```

```python
import re


def eliminate_demian_page_end_part(doc):
    target_pattern = r"\d+\s+downloaded from https\s+www\.holybooks\.com"
    cleaned_doc = re.sub(target_pattern, "\n", doc)
    
    return cleaned_doc

for doc in docs:
    doc.page_content = eliminate_demian_page_end_part(doc.page_content)

print(docs[3].page_content)
```

```python
page_content_list = []
for i in range(len(docs)):
    page_content_list.append(docs[i].page_content)

merged_docs = "\n".join(page_content_list)
merged_docs
```

```python
import tiktoken

def openai_length_function():
    encoding = tiktoken.get_encoding("o200k_base")
    
    return lambda text: len(encoding.encode(text))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100,
    length_function=openai_length_function()
)

chunks = text_splitter.split_text(merged_docs)

# 딕셔너리와 유사한(딕셔너리 타입은 아님) .metadata 안에 source, page 등 키 존재 
# .page_content에 내용 존재
# 이는 ai의 최종 결과물에도 동일하게 적용되는 것 같음
print(chunks[5])
len(chunks[60])
```

```python
from openai import OpenAI

client = OpenAI()

def get_embedding(chunks):
    embeddings = []
    for chunk in chunks:
        response = client.embeddings.create(
            input=chunk,
            model="text-embedding-3-large"
        )
        embeddings.append(response.data[0].embedding)
    return embeddings
        
embeddings = get_embedding(chunks)
```

```python
import lancedb
from lancedb.pydantic import LanceModel, Vector

db = lancedb.connect("./data")


class MetadataPydantic(LanceModel):
    page: int


class DemianPydantic(LanceModel):
    vector: Vector(3072)
    text: str
    metadata: MetadataPydantic


schemaed_data_table_list = []
for i, chunk in enumerate(chunks):
    schemaed_data_table_list.append(
        {
            "vector": embeddings[i],
            "text": chunk,
            "metadata":
            {
                "page": i+1
            }
        }
    )

table = db.create_table(
    name="demian_embedding",
    data=schemaed_data_table_list,
    schema=DemianPydantic,
    mode="overwrite"
)
```

```python
import os
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_community.vectorstores import LanceDB


llm = ChatOpenAI(
    model="gpt-5.4-mini",
    temperature=0.8,
    max_tokens=1000
)

embedding_model = OpenAIEmbeddings(
    model="text-embedding-3-large",
    dimensions=3072,
    api_key=os.getenv("OPENAI_API_KEY")
)
```

```python
from langchain_community.vectorstores import LanceDB

demian_vectorstore = LanceDB(
    connection=db,
    table_name="demian_embedding",
    embedding=embedding_model
)

demian_retriever = demian_vectorstore.as_retriever(search_kwargs={"k": 3})
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

prompt = ChatPromptTemplate.from_messages([
    (
        "system", 
        "당신은 소설 '데미안'의 깊은 심연을 꿰뚫어 보는 고결한 분석가입니다. "
        "오직 제공된 문맥(Context)만을 사용하여 답변하되, "
        "응답은 문체가 수려하며 품격과 가오가 살아있는 시적인 표현을 담아야 합니다. "
        "모르는 내용이라면 구차하게 변명하지 말고 '심연의 안개에 가려 알 수 없다'고 답하십시오."
    ),
    (
        "human",
        "### 문맥 (context)\n{context}\n\n"
        "### 질문 (question)\n{question}\n\n"
        "### 답변 (answer): "
    )
])
```

```python
chain = (
    {"context": demian_retriever, "question": RunnablePassthrough()}
    |prompt
    |llm
    |StrOutputParser()
)
```

```python
query = input()
response = chain.invoke(query)

print(response)
```

```python
from datasets import Dataset
from ragas import evaluate
from ragas.metrics import faithfulness, answer_relevancy, context_precision


data = {
    "question": [query],
    "answer": [response],
    "contexts": [[doc.page_content for doc in docs]],
    "ground_truth": []
}

dataset = Dataset.from_dict(data)

result = evaluate(
    dataset=dataset,
    metrics=[
        faithfulness,
        answer_relevancy,
        context_precision
    ],
    llm=llm,
    embedding_model=embedding_model
)

print(result)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import JsonOutputParser
from pydantic import BaseModel, Field


class QAPair(BaseModel):
    # Field는 추가적인 정보를 넘길 수 있는 기능
    question: str = Field(description="소설의 맥락을 관통하는 깊이 있는 질문")
    answer_sheet: str = Field(description="문맥에 근거한 정확하고 상세한 정답")


class QASet(BaseModel):
    # 타입힌팅의 알맹이는 대괄호로 표시한다.
    pairs: list[QAPair]


generator_llm = ChatOpenAI(model="gpt-5.4", temperature=0.1)
structured_llm = generator_llm.with_structured_output(QASet)

gen_prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "당신은 소설 '데미안'의 지문을 분석하여 RAG 평가용(RAGAS) 데이터셋을 만드는 교수입니다. "
        "제공된 문맥에서만 유추할 수 있는 정교한 질문과 정답을 생성하세요. "
        "반드시 JSON 형식으로 응답하세요."
    ),
    (
        "human",
        "### 문맥 (Context)\n{chunk}\n\n위 문맥을 바탕으로 질문-정답 쌍을 2개 생성하세요."
    )
])
```

```python
chain = gen_prompt | structured_llm
test_set= []
for i in range(0, len(chunks), 2):
    response = chain.invoke({"chunk": chunks[i]})
    test_set.append(response)
```
